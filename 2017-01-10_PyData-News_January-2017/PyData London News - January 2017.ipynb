# POTM - Grumpy <small>Python runtime in Go</small>

+ Google's Python to Go source code transcompiler and runtime 
+ Compiles Python source code to Go source code which is then compiled to native code.
+ No bytecode, no VM.
+ Project motivated by a need to improve Youtube concurrent workloads which runs front end on CPython2.7
+ Tradeoffs: no support for C extension modules, but no GIL either
+ https://github.com/google/grumpy
